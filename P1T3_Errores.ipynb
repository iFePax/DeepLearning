#Solucion de errores

##Ejercicio 1

```python
bruto = int(input('Cual es el valor bruto del producto: '))
#Se bajo (print(f_aplica_iva(bruto)))
def f_aplica_iva(bruto, iva = 0.19): #El iva se puede sacar de dos maneras (bruto*0.19)
                                      #o (bruto*(19/100))
    total_iva = bruto * iva   
    return total_iva

print(f_aplica_iva(bruto))
```

##Ejercicio 2

```python
u = list((1, 2, 3))#Las tuplas son inmutables por eso se debe usar la funcion List()
v = list((4, 5, 6))

def multiplicar_vector(u, v):
    for i in range(len(u)): #Se pone el limite en lo que es de grande el vector u
        u[i] *= v[i] #Se corrige que se llegue a una casilla con valor NULL
    return sum(u)

print(multiplicar_vector(u, v))
```

##Ejercicio 3

```python
total = 0 #Instanciar la variable total
for i in range(0, int(input("Total de números? "))): #Se cambia que s epueda ingresar enteros
  total+=int(input("Ingrese un valor: "+str(i+1)+": "))#Poner que se empiece a contar desde 1 y no desde
print(f"El total de la suma es: ",total) #Poner la f para que se pueda llamar la variable directamente
```
